```python
df = pd.read_csv('C:/Users/sriram/OneDrive/Desktop/house_sale.csv')
df
```

# DATA CLEANING

```python
df1 = df.drop(['PRT_ID','DATE_SALE','N_BEDROOM','SALE_COND','STREET','MZZONE','QS_ROOMS','QS_BATHROOM','QS_BEDROOM','QS_OVERALL','REG_FEE','COMMIS','UTILITY_AVAIL','DATE_BUILD'],axis=1)
df1
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2
```

```python
df2['AREA'].unique()
```

```python
df3 = df2.replace(['T Nagar', 'Chrompt', 'Chrmpet', 'Karapakam',
       'Ana Nagar', 'Chormpet', 'Adyr', 'Velchery', 'Ann Nagar',
       'KKNagar'],['TNagar','Chrompet','Chrompet','Karapakkam', 'Anna Nagar','Chrompet','Adyar','Velachery','Anna Nagar','KK Nagar'])
df3
```

```python
df3['BUILDTYPE'].unique()
```

```python
df4 = df3.replace(['Others','Comercial'],['Other','Commercial'])
df4
```

```python
df5 = df4.replace(['Noo'],['No'])
df5
```

```python
dummy_area = pd.get_dummies(df5['AREA'])
dummy_area
```

```python
dummy_build = pd.get_dummies(df5['BUILDTYPE'])
dummy_build
```

```python
dummy_park = pd.get_dummies(df5['PARK_FACILITY'])
dummy_park
```

```python
df6 = pd.concat([df5,dummy_area,dummy_build,dummy_park],axis=1)
df6
```

```python
df7 = df6.drop(['AREA','BUILDTYPE','Karapakkam','Other','PARK_FACILITY','No'],axis=1)
df7
```

```python
df7['park'] = df7.Yes
df7.drop(['Yes'],axis=1)
data = df7
data
```

# TRAINING MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# splitting data into training and testing
x = data.drop(['SALES_PRICE'],axis=1)
x
```

```python
y = data['SALES_PRICE']
y
```

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
x_test

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```
